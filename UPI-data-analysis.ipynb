# Monthly Performance 2021

```python
import calendar
upi_2021_df['month_name'] = upi_2021_df['Month'].apply(lambda x: calendar.month_abbr[x])
upi_2021_df
```

```python
monthly_per_2021 = upi_2021_df.groupby('Month')[['value_cr', 'month_name']].sum()
monthly_per_2021 = monthly_per_2021.reset_index()
monthly_per_2021['month_name'] = monthly_per_2021['Month'].apply(lambda x: calendar.month_abbr[x])
monthly_per_2021
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (15, 10)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.plot(monthly_per_2021.month_name, monthly_per_2021.value_cr, 'o-r')

plt.ylabel('Value (Crores)')

plt.title('2021 UPI transactins in Crores');
```

```python
print('UPI transactions amounting Rs.{:.2f} Crores in 2021'.format(monthly_per_2021['value_cr'].sum()))
```

# Marketshare 2021

```python
pie_2021_df = upi_2021_df.groupby('upi_banks')[['value_cr']].sum()
pie_2021_df = pie_2021_df.sort_values('value_cr', ascending=False)
pie_2021_df = pie_2021_df.sort_values('value_cr', ascending=False).head(5)
pie_2021_df = pie_2021_df.reset_index()
pie_2021_df['market_share'] = pie_2021_df.value_cr *100 / pie_2021_df.value_cr.sum()
pie_2021_df
```

```python
plt.figure(figsize=(8,8))

plt.title('Top 5 UPI banks marketshare in 2021')

plt.pie(pie_2021_df.market_share, labels = pie_2021_df.upi_banks, autopct = '%1.1f%%', textprops={'color':'black', 'fontsize':10});
```

# Trends UPI 2021

```python
upi_trend_2021 = upi_2021_df.groupby('Month')[['volume_mn_by_costumers', 'value_cr_by_costumers','volume_mn', 'value_cr']].sum()
```

```python
upi_trend_2021
```

```python
upi_trend_2021 = upi_trend_2021.reset_index()
```

```python
upi_trend_2021
```

```python
upi_trend_2021['month_name'] = upi_trend_2021['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
upi_trend_2021
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(upi_trend_2021.month_name, upi_trend_2021.volume_mn_by_costumers, 'o-r')
ax2.plot(upi_trend_2021.month_name, upi_trend_2021.value_cr_by_costumers, 'o-b')

ax1.set_ylabel('Volumn (Mn) by customers', color = 'r')
ax2.set_ylabel('Value (Cr) by costumers', color = 'b')

plt.title('Trends by customers in 2021');
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(upi_trend_2021.month_name, upi_trend_2021.volume_mn, 'o-r')
ax2.plot(upi_trend_2021.month_name, upi_trend_2021.value_cr, 'o-b')

ax1.set_ylabel('Volumn (Mn)', color = 'r')
ax2.set_ylabel('Value (Cr)', color = 'b')

plt.title('Trends in 2021');
```
